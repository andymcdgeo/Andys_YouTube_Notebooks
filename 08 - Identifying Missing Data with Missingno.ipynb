## Import Libraries and Data

```python
import pandas as pd
import missingno as msno

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Data/xeek_subset2.csv')
```

## Exploring Nulls with Pandas

```python
#View summary statistics of the data
df.describe()
```

```python
#View information about the data including non-null counts
df.info()
```

```python
#Summary of the number of nulls
df.isna().sum()
```

## Using Missingno

```python
#Bar Chart
msno.bar(df)
plt.savefig('bar.png', dpi=150)
```

```python
#Matrix Chart
msno.matrix(df)
plt.savefig('matrix.png', dpi=150)
```

```python
#Heatmap
msno.heatmap(df)
```

```python
#Dendrogram
msno.dendrogram(df)
```
